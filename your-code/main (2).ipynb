### 2. Linear functions

Consider the following straight lines on 2D space:  
$y_1=2x-3$  
$y_2=-3x+12$  

#### 2.1. Plot them

```python
def initialize_grid():
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # where do we want lines
    ticks = np.arange(-15, 15, 1)
    
    # draw grid
    ax.set_xticks(ticks)
    ax.set_yticks(ticks)
    ax.grid(True, which='both')
    
    # 1-1 X and Y proportion
    ax.set_aspect('equal')
    
    # X and Y axes
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    
    # set axes' limits
    ax.set_xlim([-15, 15])
    ax.set_ylim([-15, 15])
```

```python
x = np.linspace(
    start=-10, 
    stop=10, 
    num=300)

y1 = 2 * x - 3  
y2 = -3 * x + 12  
```

```python
initialize_grid()

# "-" for line type
plt.plot(x, y1, "-", markersize=10, label="2x-3")
plt.plot(x, y2, "-", markersize=2, label="-3x+12")

plt.legend()
```

#### 2.2. What are the respective slopes

```python
def slope(x1, y1, x2, y2):
    m = (y2-y1)/(x2-x1)
    return m
```

#### 3.2. Which one takes a bigger value on $x=0$? On $x=2$? On $x=100$?

```python
def value_x(x):
  y_2 = x ** 2 - 3 * x - 10
  y_3 = x ** 3 - 2 * x ** 2 + 5 * x + 1
  y_4 = x ** 4 - x ** 3 + x ** 2 - x + 1
  return y_2, y_3, y_4

def biggest(a, b, c, function_names):
  if a >= b and a >= c:
    return function_names[0]
  if b >= a and b >= c:
    return function_names[1]
  return function_names[2]

y_2, y_3, y_4 = value_x(0)
big = biggest(y_2, y_3, y_4, ['y_2', 'y_3', 'y_4'])
print('big for x=0 is', big)

y_2, y_3, y_4 = value_x(2)
big = biggest(y_2, y_3, y_4, ['y_2', 'y_3', 'y_4'])
print('big for x=2 is', big)

y_2, y_3, y_4 = value_x(100)
big = biggest(y_2, y_3, y_4, ['y_2', 'y_3', 'y_4'])
print('big for x=100 is', big)
```

### 6. Combinatorics

#### 6.1. Ten people go to a party. How many handshakes happen?

```python
def handshakes(n):
  return n * ((n - 1) / 2)
 
handshakes(10)
```

#### 6.2. My family is composed of 11 members. By christmas, one buys food, one cooks and one cleans. How many possible arrangements of buyer-cooker-cleaner are there?

```python
def variaciones(N, k):
  return (math.factorial(N)/math.factorial(N - k))
```

```python
christmas = variaciones(11, 3)
christmas
```
